- Sensor data comparative analysis between two different products used by anaesthesiologist to monitor the depth of general anesthesia. An electorde was taped to a patients forehead, in simplistic terms this electrode measures the brain activity and gives us a number (BIS) that gives us information about how deeply a patient is anesthetised. Electroncephalogram (EEG) data alongside with a few other parameters (SQI-Signal quality, EMG-electromyogram, BSR-Burst supression rate)are recorded. For the purposes of this study a program was developed that recorded not only the mentioned parameters but all biological signals that are monitored during a surgery: Oxygen saturation in blood (SpO2), EEG Frequency, Hearth rate (HR), ECG (ST II), capnography (erCO2),MAC (minimum alveolar concentration in the case inhalational anesthestic would be used). Relevant parameters (BIS, EMG, SQI, BSR)were recorded every 5 seconds, all the other parameters were recorded every second. This study compares two electrodes, a standard one and a newone, generic one. The purpose of the analysis is to assess the quality of a new generic equipment.
